Analisi del dataset del WALS (World Atlas of Language Structures).

Il dataset ha i seguenti header come struttura, il dataset è stato denormalizzato in modo da rendere più semplici certe operazioni.

```python
import sys
sys.path.append("../src")

import info
import loader
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

dataset_path = "../dataset/processed/features.csv"
    
try:
    df = loader.load_data(dataset_path)
except FileNotFoundError as e:
    print(f"Error: dataset file ({dataset_path}) not found")
    sys.exit(1)

info.print_headers(df)
```
