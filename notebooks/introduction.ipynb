Analisi del dataset del WALS (World Atlas of Language Structures).

Il dataset ha i seguenti header come struttura, il dataset non è stato denormalizzato in modo da rendere più ergonomiche certe operazioni di analisi.

```python
import sys
sys.path.append("../src")

import info
import loader

dataset_path = "../dataset/features.csv"
    
try:
    df = loader.load_data(dataset_path)
except FileNotFoundError as e:
    print(f"Error: dataset file ({dataset_path}) not found")
    sys.exit(1)

info.print_headers(df)
```

## Header riguardanti la lingua

Gli header che riguardano una cera lingua sono i seguenti:
- Family: la famiglia linguistica a cui la lingua appartiene.
- Subfamily: la sottofamiglia linguistica a cui la lingua appartiene.
- Genus: il genus a cui la lingua appartiene.
- Language_Name: il nome della lingua.
- Language_ID: il codice identificativo della lingua utilizzato dal dataset.

```python
info.print_columns(df, ["Family", "Subfamily", "Genus", "Language_Name", "Language_ID"])
```
