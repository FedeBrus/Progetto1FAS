```python
import sys
sys.path.append("../src")

import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact
import info
import loader
import stats
import plotter

dataset_path = "../dataset/processed/features.csv"
    
try:
  df = loader.load_data(dataset_path)
except FileNotFoundError as e:
  print(f"Error: dataset file ({dataset_path}) not found")
  sys.exit(1)
```

### Distribuzione geografica di alcune famiglie linguistiche

```python
def select_family(families):
  selected_family = families
  map = stats.map_points(
      df,
      "Family",
      [selected_family]
  )

  plotter.plot_points_on_map(map, "Family")
  plt.show()

interact(select_family, families=stats.get_list_of(df, "Family"));
```

```python
def select_family(genii):
  selected_genus = genii 
  map = stats.map_points(
      df,
      "Genus",
      [selected_genus]
  )

  plotter.plot_points_on_map(map, "Genus")
  plt.show()

interact(select_family, genii=stats.get_list_of(df, "Genus"));
```

### Distribuzione geografica di alcuni valori di parametri

```python
def select_feature(features):
  feature_id = df[df["Parameter_Name"] == features]["Parameter_ID"].iloc[0]
  map = stats.map_points(
      df,
      "Parameter_ID",
      [feature_id]
  )

  plotter.plot_points_on_map(map, "Code_Name")
  plt.show()

interact(select_feature, features=stats.get_list_of(df, "Parameter_Name"));
```

### Distribuzione geografica dei valori di un parametro filtrato per famiglia

```python
def select_family_and_feature(features, families):
    qdf = stats.filter(df, "Family", [families])

    rows = qdf[qdf["Parameter_Name"] == features]

    # No feature for that family, i.e. empty result
    if (rows.empty):
      return

    feature_id = rows["Parameter_ID"].iloc[0]
     
    map_data = stats.map_points(
        qdf,
        "Parameter_ID",
        [feature_id]
    )

    plotter.plot_points_on_map(map_data, "Code_Name")
    plt.show()

interact(
    select_family_and_feature, 
    features=stats.get_list_of(df, "Parameter_Name"),
    families=stats.get_list_of(df, "Family")
);
```

```python
def select_family_and_feature(features, genii):
    qdf = stats.filter(df, "Genus", [genii])

    rows = qdf[qdf["Parameter_Name"] == features]

    # No feature for that family, i.e. empty result
    if (rows.empty):
      return

    feature_id = rows["Parameter_ID"].iloc[0]
     
    map_data = stats.map_points(
        qdf,
        "Parameter_ID",
        [feature_id]
    )

    plotter.plot_points_on_map(map_data, "Code_Name")
    plt.show()

interact(
    select_family_and_feature, 
    features=stats.get_list_of(df, "Parameter_Name"),
    genii=stats.get_list_of(df, "Genus")
);
```
