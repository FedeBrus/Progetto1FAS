```python
import sys
sys.path.append("../src")

import pandas as pd
import matplotlib.pyplot as plt
import info
import loader
import stats
import plotter

dataset_path = "../dataset/features.csv"
    
try:
    df = loader.load_data(dataset_path)
except FileNotFoundError as e:
    print(f"Error: dataset file ({dataset_path}) not found")
    sys.exit(1)
```

### Distribuzione geografica di alcune famiglie linguistiche

```python
map = stats.map_points(
    df,
    "Family",
    ["Dravidian"]
)

plotter.plot_density_on_map(map, "Family")
plt.show()
```

### Distribuzione geografica di alcuni valori di parametri

```python
feature_id = "1A"

map = stats.map_points(
    df,
    "Parameter_ID",
    [feature_id]
)

plotter.plot_points_on_map(map, "Code_Name")
plt.show()
```

### Distribuzione geografica dei valori di un parametro filtrato per famiglia

```python
feature_id = "23A"
family_name = "Altaic"

qdf = df.copy()
qdf = stats.filter(qdf, "Family", [family_name])

map = stats.map_points(
    qdf,
    "Parameter_ID",
    [feature_id]
)

plotter.plot_points_on_map(map, "Code_Name")
plt.show()
```
